# Observed column review rating has got 37 null values 
As review rating is different for each category 
Hence we find median review rating for each category and replace it with the null values

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#Renaming the columns for easier reference
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Creating age_group column
#dividing the age column into 4 groups to understand shopping patterns wrt to each age group
labels = ['Young Adult','Adult','Middle-Aged','Senior-Citizens']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
#Read first 10 rows of age and age group 
df[['age','age_group']].head(10)
```

```python
# Create purchase_frequency column
#As in the given data the frequency of purchasing is given in text format
#and is harder to work with we transform it numerical data

#Created a dictionary and used it to map
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Bi-Weekly' : 14,
    'Quarterly' : 90,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
